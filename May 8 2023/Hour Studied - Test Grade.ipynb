```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
features = np.array(
    [
        [
            2, 3, 4, 5, 6, 7, 8, 9, 10
        ]
    ]
).T
features
```

```python
targets = np.array(
    [
        [
            57, 66, 73, 76, 79, 81, 90, 96, 100
        ]
    ]
).T
targets
```

```python
plt.scatter(features, targets)
```

```python
plt.plot(features, targets, 'ro')
plt.axis([2, 10, 60, 100])
plt.xlabel('Hour Studied')
plt.ylabel('Test Grade')
plt.show()
```

```python
# Fit the model by Linear Regression
classifier = linear_model.LinearRegression()
classifier.fit(features, targets)
```

```python
# y = w0 + w1 * x
# target = coef_ * feature + intercept_

w = classifier.coef_[0]
b = classifier.intercept_

print('coef w = ', w) # Hệ số
print('intercept_ b = ', b) # Sai số
```

```python
# Vẽ đường thẳng đi qua các điểm dữ liệu

# Fitting line
x0 = np.linspace(2, 9, 2, endpoint=True)
y0 = (w * x0) + b

# Visualize data
plt.plot(features, targets, 'ro')
plt.plot(x0, y0)
plt.axis([2, 10, 60, 100])
plt.xlabel('Hour Studied')
plt.ylabel('Test Grade')
plt.show()
```

```python
# Dự đoán điểm số cho 1 học sinh học 7 giờ
print("Hours = 7, Predicted Test Grade = ", classifier.predict([[7]]))
```
